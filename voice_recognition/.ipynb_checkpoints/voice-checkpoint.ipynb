```python
import sounddevice as sd
import numpy as np
import queue
import threading
import ipywidgets as widgets
from IPython.display import display, Audio
import whisper
import warnings

warnings.simplefilter(action='ignore', category=FutureWarning)

model = whisper.load_model("base")

q = queue.Queue()
recording = threading.Event()

audio_data = np.array([], dtype=np.float32)

def start_recording():
    global audio_data
    recording.set()
    audio_data = np.array([], dtype=np.float32)

    def callback(indata, frames, time, status):
        if recording.is_set():
            q.put(indata.copy())

    with sd.InputStream(callback=callback, channels=1, samplerate=16000, dtype='float32'):
        while recording.is_set():
            sd.sleep(100)

# Arrêter l'enregistrement et récupérer les données
def stop_recording_and_save():
    global audio_data
    recording.clear()
    audio_list = []
    while not q.empty():
        audio_chunk = q.get()
        audio_list.append(audio_chunk)
    if audio_list:
        # Concatenation et aplatissage des données audio
        audio_data = np.concatenate(audio_list, axis=0).flatten()
        print("Enregistrement terminé.")

# Transcription
def transcribe_audio():
    global audio_data
    if audio_data.size > 0:
        result = model.transcribe(audio_data, fp16=False, temperature=0.7)
        print("Transcription : ", result['text'])
        print (result)
    else:
        print("Aucune donnée audio à transcrire.")

# Lecture de l'audio avec barre de progression
def play_audio(b):
    global audio_data
    if audio_data.size > 0:
        # Normaliser les données pour être dans la plage [-1, 1]
        max_val = np.max(np.abs(audio_data))
        if max_val > 0:
            audio_norm = audio_data / max_val
        else:
            audio_norm = audio_data
        audio_norm = audio_norm.astype(np.float32)
        audio_widget = Audio(audio_norm, rate=16000, autoplay=True)

        display(audio_widget)
    else:
        print("Aucune donnée audio à lire.")

# Widgets de contrôle
start_button = widgets.Button(description="Start Recording", button_style='success')
stop_button = widgets.Button(description="Stop Recording", button_style='danger')
transcribe_button = widgets.Button(description="Transcribe Audio", button_style='info')
play_button = widgets.Button(description="Play Audio", button_style='primary')

start_button.on_click(lambda x: threading.Thread(target=start_recording, daemon=True).start())
stop_button.on_click(lambda x: stop_recording_and_save())
transcribe_button.on_click(lambda x: transcribe_audio())
play_button.on_click(play_audio)

control_box = widgets.VBox([
    widgets.HBox([start_button, stop_button, transcribe_button, play_button])
])

display(control_box)
```
